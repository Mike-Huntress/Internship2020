```python
dl = DataLib("SignalData")
```

```python
# INDICATOR CONSIDERATION 1

# relative stock and bond performance
# during relatively higher equity growth, people will tend to favor equity investment 

WEEKDAYS_IN_YEAR = 261
QUARTER_YEAR = 65
BondReturns = dl.pull('BondRetIdx/LocalFX')
Equities = dl.pull('EquityPrices')

# GENERATE TO SEE LINKAGE
Bond_semiannual_return = BondReturns.pct_change(QUARTER_YEAR, fill_method = None)
Bond_semiannual_return = Bond_semiannual_return.shift(-QUARTER_YEAR) # this will be "future returns" (starting today) instead of past returns (starting 6 months ago)

# calculate the rolling average annual growth of stocks and bonds, drop na so starts 1 year later
average_annual_growth_bond = BondReturns.pct_change(WEEKDAYS_IN_YEAR, fill_method = None)
average_annual_growth_equity = Equities.pct_change(WEEKDAYS_IN_YEAR, fill_method = None)
rmag_bond = average_annual_growth_bond.rolling(len(average_annual_growth_bond), min_periods=1).mean()
rmag_equity = average_annual_growth_equity.rolling(len(average_annual_growth_equity), min_periods=1).mean()

# copy dates into a new indicator data frame
ind_1_return_diff = rmag_bond.copy()
ind_1_return_diff[ind_1_return_diff != 0] = 0

# iterate and calculate the indicator
for country, series in ind_1_return_diff.iteritems():
    for date, _ in series.iteritems():
        recent_bond_growth = average_annual_growth_bond[country][date]
        recent_equity_growth = average_annual_growth_equity[country][date]
        
        mean_bond_growth = rmag_bond[country][date]
        mean_equity_growth = rmag_equity[country][date]
        
        bond_growth_ratio = recent_bond_growth/mean_bond_growth - 1
        equity_growth_ratio = recent_equity_growth/mean_equity_growth - 1
        
        # take the difference of the growth ratios (limit to between -1 and 1); may be NaN
#         ind_1_return_diff[country][date] = max(min((bond_growth_ratio - equity_growth_ratio),1),-1)
        ind_1_return_diff[country][date] = bond_growth_ratio - equity_growth_ratio

country = 'AUS'
BondReturns[country].plot()
plt.show()


fig, ax1 = plt.subplots()
fig.set_figheight(10)
fig.set_figwidth(20)
ax2 = ax1.twinx()
ax1_label = 'Bond Quarterly Future Return'
ax1.plot(Bond_semiannual_return[country].to_timestamp(), label = ax1_label)
ax1.set_ylabel(ax1_label)
ax1.legend()
ax2_label = 'Equity Bond Return Difference'
ax2.plot(ind_1_return_diff[country].to_timestamp(), color = 'tab:red', label = ax2_label)
ax2.set_ylabel(ax2_label)
ax2.legend()
plt.title('Bond Quarterly Forward Return and Equity Bond Return Difference')
handles, labels = ax1.get_legend_handles_labels()
fig.legend(handles, labels, loc='upper right')
plt.show()
```

```python
# INDICATOR 2 Real GDP and Inflation Growth

WEEKDAYS_IN_YEAR = 261
QUARTER_YEAR = 65

BondReturnIndex = dl.pull('BondRetIdx/LocalFX')

# GENERATE TO SEE LINKAGE
Bond_semiannual_return = BondReturnIndex.pct_change(QUARTER_YEAR, fill_method = None)
Bond_semiannual_return = Bond_semiannual_return.shift(-QUARTER_YEAR) # this will be "future returns" (starting today) instead of past returns (starting 6 months ago)


gdpNom = dl.pull("GDP/Nominal")
gdpReal = dl.pull("GDP/Real")
inflation = dl.pull("CoreCPI/SA")

# graph of the GDP growth vs. future quarter bond returns
rgdp_growth = gdpReal.pct_change()
inflation_growth = inflation.pct_change()

# rolling 5-year average of the inflation growth
rmg_inflation = inflation_growth.rolling(5*12, min_periods=1).mean()

# rolling 10-year average of RGDP growth
rmg_rgdp = rgdp_growth.rolling(10*4, min_periods=1).mean()

inflation_growth_mean_diff = inflation_growth - rmg_inflation
rgdp_growth_mean_diff = rgdp_growth - rmg_rgdp

# indicator 2 about business cycles
# inflation growth higher than normal, gdp growth higher than normal, go short, since expansionary, IR will increase
# inflation growth lower than normal, gdp growth higher than normal, production increasing, economy growing, go long, 
# but defer to relative performance of equity/bonds (ind_1)
# inflation growth higher than normal, gdp growth lower than normal, stagflation, go short, IR will increase
# inflation growth lower than normal, gdp growth lower than normal, recession, go long, IR will decrease

# copy dates into a new indicator data frame
ind_2_gdp_infla = inflation_growth_mean_diff.copy()
ind_2_gdp_infla[ind_2_gdp_infla != 0] = 0

# iterate and calculate the indicator
for country, series in ind_2_gdp_infla.iteritems():
    for date, _ in series.iteritems():
        inflation_diff = inflation_growth_mean_diff[country][date]
        gdp_diff = rgdp_growth_mean_diff[country][date]
        
        # position opposite of the inflation and gdp (i.e. both above average -> go short=negative, expect higher IR)
        if math.isnan(inflation_diff) and math.isnan(gdp_diff):
            ind_2_gdp_infla[country][date] = float('nan')
        elif math.isnan(inflation_diff):
            ind_2_gdp_infla[country][date] = -1 * (gdp_diff)
        elif math.isnan(gdp_diff):
            ind_2_gdp_infla[country][date] = -1 * (inflation_diff)
        else:
            ind_2_gdp_infla[country][date] = -1 * (inflation_diff + gdp_diff)
    

country = 'JPN'

fig, ax1 = plt.subplots()
fig.set_figheight(10)
fig.set_figwidth(20)
ax2 = ax1.twinx()
ax1_label = 'Bond Quarterly Future Return'
ax1.plot(Bond_semiannual_return[country].to_timestamp(), label = ax1_label)
ax1.set_ylabel(ax1_label)
ax1.legend()
ax2_label = 'Real GDP/Inflation growth'
ax2.plot(ind_2_gdp_infla[country].to_timestamp(), color = 'tab:red', label = ax2_label)
ax2.set_ylabel(ax2_label)
ax2.legend()
plt.title('Bond Quarterly Forward Return and Real GDP/Inflation Growth')
handles, labels = ax1.get_legend_handles_labels()
fig.legend(handles, labels, loc='upper right')
plt.show()

# print(ind_2_gdp_infla[country])
# graph of the GDP
# plt.figure(figsize=(20,5))
# inflation[country].plot()
# plt.title('gdp real')
# plt.show()
```

```python
# SIGNAL CREATION FUNCTION

# This is a function that takes in three data streams and produces a signal for each date
# Assume that each data stream is for one country only (call this function for each country if multiple countries)
# Returns a vector that is a signal x in [-1,1] for each date where x is a position for the next day

def generate_signal(ind_1, ind_2 = None, ind_3 = None):
    if ind_2 is None:
        return ind_1.dropna()
    else:
        return ind_2.dropna()

    
```

```python
# TEST RETURNS

def test_returns(country, signal):
    bond_returns = dl.pull('BondRetIdx/LocalFX')
    
    # upsample in case dates are not by business days
    upsampled_signal = signal.resample('1B').ffill()
    
    earliest_date = upsampled_signal.index.min()
    latest_date = bond_returns.index.max()
    
    lagged_bond_returns = bond_returns[country].pct_change().shift(-1)[earliest_date:]
    daily_returns =  lagged_bond_returns * upsampled_signal
    
    sum_our_returns = daily_returns.cumsum()
    sum_bond_returns = lagged_bond_returns.cumsum()
    
    plt.figure(figsize=(20,5))
    sum_our_returns.plot()
    sum_bond_returns.plot()
    plt.title('Cummulative Returns')
    plt.legend(['Our ' + country + ' Cummulative Returns', country + ' Cummulative Bond Returns'])
    plt.show()
    
    print('Active Money Made: ' + str(sum_our_returns[latest_date - BDay(1)]))
    print('Passive Money Made: ' + str(sum_bond_returns[latest_date - BDay(1)]))
    print('Earliest Date: ' + str(earliest_date))
    
def show_signals(signal):
    plt.figure(figsize=(20,1))
    signal.plot()
    plt.title('Signals Over Time')
    plt.show()
```

```python
# SEE RESULTS

all_countries = ['AUS','DEU','CAN','ESP','FRA','ITA','JPN','CHE','GBR','USA']
# all_countries = ['AUS','DEU','CAN']

for country in all_countries:
    indicator_1 = ind_1_return_diff[country]
    indicator_2 = ind_2_gdp_infla[country]
    signal = generate_signal(indicator_1, indicator_2)
    test_returns(country, signal)
    show_signals(signal)
    
```
