```python
import sys
sys.path.append('..')

import matplotlib.pyplot as plt
import MySignal as sig

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
countries=['USA']
world_countries = ['USA', 'CAN', 'AUS', 'GBR', 'CHE']
beg_date, end_date = ('1995-01-01', '2005-12-31')
```

```python
equities_growth_diff = sig.StandardizedRollingGrowthDiff(series_names=['BondRetIdx/LocalFX'],
                                                         weights_dict={'BondRetIdx/LocalFX': -1},
                                                         world=world_countries,
                                                         time_window=261*5,
                                                         n_rolling_periods=20)
equities_growth_diff.plot_indicator_returns_diff(world_countries, beg_date, end_date)
```

```python
equities_growth_diff_signal = sig.SigmoidNormalizerDiff(indicators_dict={'equities_diff': equities_growth_diff},
                                                        weights_dict={'equities_diff': -1},
                                                        world=world_countries)
equities_growth_diff_signal.plot_signal_index_returns(world_countries, beg_date, end_date)
equities_growth_diff_signal.get_metrics(world_countries, beg_date, end_date)
```

```python
gdp_growth_diff = sig.StandardizedRollingGrowthDiff(series_names = ['GDP/Real'],
                                                    weights_dict = {'GDP/Real': 1},
                                                    world = world_countries,
                                                    time_window=261*5
                                                   )
gdp_growth_diff.plot_indicator_returns_diff(world_countries, beg_date, end_date)
```

```python
gdp_growth_diff_signal = sig.SigmoidNormalizerDiff(indicators_dict={'gdp_diff': gdp_growth_diff},
                                                   weights_dict={'gdp_diff': -1},
                                                   world=world_countries)
gdp_growth_diff_signal.plot_signal_index_returns(world_countries, beg_date, end_date)
gdp_growth_diff_signal.get_metrics(world_countries, beg_date, end_date)
```

```python
equities_gdp_growth_diff_signal = sig.SigmoidNormalizerDiff(indicators_dict={'gdp_diff': gdp_growth_diff,
                                                                             'equities_diff': equities_growth_diff},
                                                            weights_dict={'gdp_diff': -1,
                                                                          'equities_diff': -1},
                                                            world = world_countries)
equities_gdp_growth_diff_signal.plot_signal_index_returns(world_countries, beg_date, end_date)
equities_gdp_growth_diff_signal.get_metrics(world_countries, beg_date, end_date)
```

```python
beta_equities_gdp_growth_diff_signal = sig.SigmoidNormalizerDiff(indicators_dict={'gdp_diff': gdp_growth_diff,
                                                                                  'equities_diff': equities_growth_diff},
                                                                 weights_dict={'gdp_diff': -1,
                                                                               'equities_diff': -1},
                                                                 world=world_countries,
                                                                 beta_overlay=0.6)
beta_equities_gdp_growth_diff_signal.plot_signal_index_returns(world_countries, beg_date, end_date)
beta_equities_gdp_growth_diff_signal.get_metrics(world_countries, beg_date, end_date)
```
