## GDP Growth

```python
GDP_growth_indicator = sig.StandardizedGrowth(series_names=['GDP/Real'],
                                          weights_dict={'GDP/Real': 1})
GDP_growth_signal = sig.SigmoidNormalizerSignal(indicators_dict={'GDP_growth': GDP_growth_indicator},
                                            weights_dict={'GDP_growth': -1})
GDP_growth_signal.plot_signal_index_returns(['USA'], '1995-01-01', '2005-12-31')
GDP_growth_signal.plot_signal_sharpe(['USA'], '1995-01-01', '2005-12-31')
GDP_growth_indicator.plot_indicator_returns(['USA'], '1995-01-01', '2005-12-31')
```
