```python
import sys
sys.path.append('..')

import matplotlib.pyplot as plt
import MySignal as sig

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from DataIOUtilities.DataLib import DataLib
dl = DataLib("SignalData")
```

```python
real_GDP = dl.pull("GDP/Real").resample('1Q').last()
bond_returns_index = dl.pull("BondRetIdx/LocalFX").resample('1Q').last()
short_rates = dl.pull("ShortRates").resample('1Q').last()
long_rates = dl.pull("LongRates").resample('1Q').last()
curve_height = dl.pull("CurveHeight").resample('1Q').last()
yield_curve = dl.pull("YieldCurve").resample('1Q').last()

country = 'USA'
beg_date, end_date = ('1980Q1', '2015Q3')

real_GDP = real_GDP.loc[pd.Period(beg_date):pd.Period(end_date)]
bond_returns_index = bond_returns_index.loc[pd.Period(beg_date):pd.Period(end_date)]
short_rates = short_rates.loc[pd.Period(beg_date):pd.Period(end_date)]
long_rates = long_rates.loc[pd.Period(beg_date):pd.Period(end_date)]
curve_height = curve_height.loc[pd.Period(beg_date):pd.Period(end_date)]
yield_curve = yield_curve.loc[pd.Period(beg_date):pd.Period(end_date)]
```

```python
real_GDP_growth = sig.standardize(sig.annualize(real_GDP.pct_change()))
bond_returns_growth = sig.standardize(sig.annualize(bond_returns_index.pct_change()))

fig, ax1 = plt.subplots(figsize=(9,6))
ax2 = ax1.twinx()

ax1.plot(real_GDP_growth[country].to_timestamp(), color='black', label='GDP')
ax1.set_ylabel('Growth (%)')
ax1.legend(loc=2)

#ax2.plot(bond_returns_growth[country].to_timestamp(), color='blue', label='Bond Index')
#ax2.set_ylabel('Returns (%)')
#ax2.legend(loc=1)

plt.xlabel('Date')

plt.title('{} GDP Growth vs Bond Returns vs Strategy Returns'.format(country))
plt.show()
```

```python
real_GDP_growth = sig.standardize(sig.annualize(real_GDP.pct_change()))

short_rates_stand = sig.standardize(short_rates)
long_rates_stand = sig.standardize(long_rates)

fig, ax1 = plt.subplots(figsize=(9,6))
ax2 = ax1.twinx()

ax1.plot(real_GDP_growth[country].to_timestamp(), color='black', label='GDP')
ax1.set_ylabel('Growth')
ax1.legend(loc=2)

ax2.plot(short_rates_stand[country].to_timestamp(), color='blue', label='Short Rates')
ax2.plot(long_rates_stand[country].to_timestamp(), color='red', label='Long Rates')
ax2.set_ylabel('Interest Rates')
ax2.legend(loc=1)

plt.show()
```
