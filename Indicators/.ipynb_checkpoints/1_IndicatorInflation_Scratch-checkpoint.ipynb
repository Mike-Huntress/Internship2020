```python
# Generate signal for M1

M1 = dl.pull("M2/inUSD")

all_countries = M1.columns

# Generate 3M Series
M1_3M_PctChange_Unanualized = M1.pct_change(3, fill_method = None).dropna(how='all')
M1_3M_PctChange = (M1_3M_PctChange_Unanualized + 1)**(12/3) - 1

# Generate 1Y series
M1_1Y_PctChange = M1.pct_change(12, fill_method = None).dropna(how='all')


M1_3M_Minus_1Y_Change = M1_3M_PctChange - M1_1Y_PctChange


# Generate Actual Inflation measure
CoreCPI = dl.pull("CoreCPI/SA")
Inflation_Unannualized = CoreCPI.pct_change(1, fill_method = None).dropna(how='all')
Inflation = (Inflation_Unannualized + 1)**(12/1) - 1


CoreCPI.plot()
plt.xlabel('Date')
plt.ylabel('CPI')
plt.title('CPI By Country')
plt.show(block = False)

Inflation.plot()
plt.xlabel('Date')
plt.ylabel('Inflation (%)')
plt.title('Inflation By Country')
plt.show(block = False)

for country in all_countries:
    series0 = [M1_3M_PctChange[country], "M1_3M_PctChange", "M1_3M_PctChange"]
    series1 = [M1_1Y_PctChange[country], "M1_1Y_PctChange", "M1_1Y_PctChange"]
    series2 = [M1_3M_Minus_1Y_Change[country], "M1_3M_Minus_1Y_Change", "M1_3M_Minus_1Y_Change"]

    Util.GraphDifferentAxes([series0, series1, series2], "Linkages for " + country)



for country in all_countries:
    series0 = [M1_3M_Minus_1Y_Change[country], "M1_3M_Minus_1Y_Change", "M1_3M_Minus_1Y_Change"]
    series1 = [Inflation[country], "Inflation", "Inflation"]

    Util.GraphDifferentAxes([series0, series1], "Linkages for " + country)
```
