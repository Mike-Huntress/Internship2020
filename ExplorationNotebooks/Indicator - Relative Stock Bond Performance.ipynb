**Approach**:
- Core principles: The purpose of this indicator is to track the returns on equities vs. bonds (on a risk-adjusted basis). This will act as a proxy for the opportunity cost investors are faced with. As the risk-adjusted returns for equities rise, bond yields will be expected to rise to compete for capital. 
- Systemization:
    - For each country: Measure the difference between risk-adjusted returns for equities and bonds. Smooth over a short window (perhaps 20 days) to remove any relatively "random" fluctuations. This will act as a normalization over time. 
    - Calculate z-scores to find the relative ranking of each country's eqwuity/bond outperformance in compared to the global average at that point in time. This will seek to put each country's movement and our resulting positions in perspective to the global market.
