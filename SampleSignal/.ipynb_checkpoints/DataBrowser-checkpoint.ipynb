```python
# Part 2
WEEKDAYS_IN_YEAR = 261

bondReturnIdx_locFXRolling = bondReturnIdx_locFX.rolling(3 * WEEKDAYS_IN_YEAR, min_periods = WEEKDAYS_IN_YEAR).mean()

bondReturnIdx_locFXRolling.plot()
plt.title('3-year rolling Bond Return')
plt.show()



print(bondReturnIdx_locFXRolling.dropna(thresh=10))

# In general, we can see that Spain strongly outperformed the rest of the world from about 2014 to 2016.  
# This is perhaps due to the fact that Spain was hit hard during the Euro debt crisis and recovered
# meaning that interest rates fell, driving bond prices upward.
# A country that strongly underperformed the world was the USA from approximately from 2015 to 2020, perhaps because
# as the rest of the world was lowering interest rates, the US increased them.  This also explains the recent uptick
# in US sovereign bonds now, because the fed decreased rates just as the pandemic hit to near zero levels.

bondReturnIdx_locFXDownSample = bondReturnIdx_locFX.resample('1Y').mean()
print(bondReturnIdx_locFXDownSample)
```

```python
# Part 3
# Unhedged bond returns
# fxVsUSD and bondReturnIdx_locFX
# countryList = ['USA', 'AUS', 'JPN', 'CAN', 'CHE', 'GBR', 'ESP', 'FRA', 'ITA', 'DEU']
bondReturnIdx_locFXDownSampled = bondReturnIdx_locFX.resample('1M').mean()

bondReturnIdx_valid_list = []
# fxVsUSD_valid = dict()
fxVsUSD_valid_list = []
for country in bondReturnIdx_locFXDownSampled.columns:
    # bondReturnIdx_valid[country] = bondReturnIdx_locFXDownSampled.loc[bondReturnIdx_locFXDownSampled[country].first_valid_index(), country]
    bondReturnIdx_valid_list.append(bondReturnIdx_locFXDownSampled.loc[bondReturnIdx_locFXDownSampled[country].first_valid_index(), country])
    # fxVsUSD_valid[country] = fxVsUSD.loc[fxVsUSD[country].first_valid_index(), country]
    fxVsUSD_valid_list.append(fxVsUSD.loc[fxVsUSD[country].first_valid_index(), country])



bondReturnIdx_locFXDownSampled_relative = bondReturnIdx_locFXDownSampled/bondReturnIdx_valid_list
fxVsUSD_relative = fxVsUSD/fxVsUSD_valid_list

# print(bondReturnIdx_locFXDownSampled_relative)
# print('##################')
# print(fxVsUSD_relative)

bondReturnIdx_base = bondReturnIdx_locFXDownSampled_relative.mul(fxVsUSD_relative)


bondReturnIdx_base.plot()
plt.title('Unhedged Bond Returns')
plt.show()

# Hedged bond returns
bondReturnIdx_locFXDownSampled.plot()
plt.title('Hedged Bond Returns')
plt.show()



    
```

```python
# Part 4
# Average Annual Return of a 60/40 portfolio
# bondReturnIdx_locFX and equityPrices


# print(bondReturnIdx_locFX)
# print('#########')
# print(equityPrices)

# print(equityPrices['USA'])
# print('###########')
# print(bondReturnIdx_locFX['USA'])
# print('###########')

# The US 60/40 Portfolio
us_60_40 = 0.6*equityPrices['USA'] + 0.4*bondReturnIdx_locFX['USA']
# print('#######us_60_40#######')
# print(us_60_40)

# We will pick an equally weighted global bond portfolio for all the years that there is information for every country
globalBondPortfolio = bondReturnIdx_locFX.dropna(thresh=10).mean(axis=1)
# print('#######globalBondPortfolio#######')
# print(globalBondPortfolio)
# print(type(globalBondPortfolio))

# print('########')
# print(bondReturnIdx_locFX)

# print(type(us_60_40))

# Select the dates from the us_60_40 Series for which there are values in the globalBondPortfolio
us_60_40_modified = us_60_40['1991-03-29':]
# print('#######us_60_40_modified#######')
# print(us_60_40_modified)


globalBondPortfolio_1year = globalBondPortfolio.resample('1Y').mean()
us_60_40_modified_1year = us_60_40_modified.resample('1Y').mean()

# print('#######globalBondPortfolio_1year#######')
# print(globalBondPortfolio_1year)

# print('#######us_60_40_modified_1year#######')
# print(us_60_40_modified_1year)

# US 60/40 Portfolio
us_60_40_AnnualReturn = us_60_40_modified_1year.pct_change()
us_60_40_AverageAnnualReturn = us_60_40_AnnualReturn.mean()
us_60_40_VarianceAnnualReturn = us_60_40_AnnualReturn.var()
# Calculation of the Sharpe Ratio assuming a risk free return of 0
us_60_40_SharpeRatio = us_60_40_AverageAnnualReturn / us_60_40_VarianceAnnualReturn**(1/2)

print('#######us_60_40_AnnualReturn#######')
print(us_60_40_AnnualReturn)

print('#######us_60_40_AverageAnnualReturn#######')
print(us_60_40_AverageAnnualReturn)

print('#######us_60_40_VarianceAnnualReturn#######')
print(us_60_40_VarianceAnnualReturn)

print('#######us_60_40_SharpeRatio#######')
print(us_60_40_SharpeRatio)

# Equally Weighted Global Bond Portfolio
globalBondPortfolio_AnnualReturn = globalBondPortfolio_1year.pct_change()
globalBondPortfolio_AverageAnnualReturn = globalBondPortfolio_AnnualReturn.mean()
globalBondPortfolio_VarianceAnnualReturn = globalBondPortfolio_AnnualReturn.var()
# Calculation of the Sharpe Ratio assuming a risk free return of 0
globalBondPortfolio_SharpeRatio = globalBondPortfolio_AverageAnnualReturn / globalBondPortfolio_VarianceAnnualReturn**(1/2)

print('#######globalBondPortfolio_AnnualReturn#######')
print(globalBondPortfolio_AnnualReturn)

print('#######globalBondPortfolio_AverageAnnualReturn#######')
print(globalBondPortfolio_AverageAnnualReturn)

print('#######globalBondPortfolio_VarianceAnnualReturn#######')
print(globalBondPortfolio_VarianceAnnualReturn)

print('#######globalBondPortfolio_SharpeRatio#######')
print(globalBondPortfolio_SharpeRatio)


# Overall, it does seem like an equally weighted global developed world bond portfolio has a much
# higher Sharpe Ratio, implying that the returns for the given level of risk are superior
```
