```python
def firstvalid(x):
    return x[x.first_valid_index()]

a = bondReturnIdx_locFX.copy()
initial_prices = a.apply(firstvalid, axis=0)
initial_prices
```

```python
#hedged returns (in each country's currency):
#bondReturnIdx_locFX.plot()

hedgedreturns = bondReturnIdx_locFX / initial_prices
hedgedreturns.plot()
plt.show()

#hedged returns (in dollars):
b = fxVsUSD.copy()
b.loc[:, 'USA'] = 1
initial_exchange_rates = b.apply(firstvalid, axis=0)
#initial_exchange_rates
ey = bondReturnIdx_locFX.resample('M').last()
bond_unhedged_returns =  ey * fxVsUSD / initial_exchange_rates/ initial_prices
bond_unhedged_returns.plot()

a = fxVsUSD.drop(['ITA', "JPN", "ESP", "FRA"], axis=1)
a.plot()
```

```python
#hedged returns (in each country's currency):
#bondReturnIdx_locFX.plot()

hedgedreturns = bondReturnIdx_locFX.loc[:,"GBR"]  / initial_prices.loc["GBR"]
#hedgedreturns.plot()

a = hedgedreturns.loc['1990-1-1':'1994-1-1']
a.plot()
#hedged returns (in dollars):
b = fxVsUSD.copy()
b.loc[:, 'USA'] = 1
initial_exchange_rates = b.apply(firstvalid, axis=0).loc["GBR"]
#initial_exchange_rates
ey = bondReturnIdx_locFX.resample('M').last().loc[:, "GBR"]
bond_unhedged_returns =  ey * fxVsUSD.loc[:,"GBR"] / initial_exchange_rates / initial_prices.loc["GBR"]
b = bond_unhedged_returns.loc['1990-1-1':'1994-1-1']
#bond_unhedged_returns.plot()
b.plot()
plt.show()

a = fxVsUSD.loc['1990-1-1':'1994-1-1',"GBR"]
a.plot()
```

```python
#sharpe stuff
#calculate average annual return and volatility of 60/40 portfolio
equities = dsPuller.ds_country_pull(lambda x: MSCIEquityDictionary[x], start_date, 'MSPI', 'Y')
bonds = dsPuller.ds_country_pull(lambda x: f'BM{x}10Y', start_date, 'RI', 'Y')
equities_annual_return = equities.pct_change(1)
bonds_annual_return = bonds.pct_change(1)

```

```python
sixty = .6*equities_annual_return
forty = .4*bonds_annual_return
sixty_forty = sixty.add(forty)
sixty_forty_mean = sixty_forty.mean()
sixty_forty_std = sixty_forty.std()
sixty_forty_sharpe = sixty_forty_mean / sixty_forty_std
sixty_forty_sharpe
```

```python
global_bonds = bonds.mean(axis=1)
global_bond_returns = global_bonds.pct_change(1)
global_bond_returns
old_sixty_forty = .8*sixty_forty
new_global_bonds = .2*global_bond_returns
new_portfolio = old_sixty_forty.add(new_global_bonds, axis=0)
new_portfolio
new_mean = new_portfolio.mean()
new_std = new_portfolio.std()
new_sharpe = new_mean / new_std
new_sharpe
```
